```python
def sp_noise(img):
 
    # Getting the dimensions of the image
    row , col = img.shape
     
    # Randomly pick some pixels in the
    # image for coloring them white
    # Pick a random number between 300 and 10000
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
       
        # Pick a random y coordinate
        y_coord=random.randint(0, row - 1)
         
        # Pick a random x coordinate
        x_coord=random.randint(0, col - 1)
         
        # Color that pixel to white
        img[y_coord][x_coord] = 255
         
    # Randomly pick some pixels in
    # the image for coloring them black
    # Pick a random number between 300 and 10000
    number_of_pixels = random.randint(300 , 10000)
    for i in range(number_of_pixels):
       
        # Pick a random y coordinate
        y_coord=random.randint(0, row - 1)
         
        # Pick a random x coordinate
        x_coord=random.randint(0, col - 1)
         
        # Color that pixel to black
        img[y_coord][x_coord] = 0
         
    return img
```

```python
for img_class in targetnames:
    img_list = os.listdir(r'Species/Training_Set/' + img_class)
    
    for file_name in img_list:
        source = os.path.join(r'Species/Training_Set/' + img_class, file_name)
        target = os.path.join(r'Species/Adding_Noise_Salt&Pepper/' + img_class, file_name)
        print(source)

        image = cv2.imread(source,cv2.IMREAD_GRAYSCALE)
        #kalau mau hilangin grayscale nya 
#         image = cv2.imread(source)
        #berarti kasus nya di test harus di noise kan juga
        noise_img = sp_noise(image)
        cv2.imwrite(target, noise_img)
```
